#### **10. Visualize differences in data distribution between city and resort hotels?**

````python
numeric_features = ['lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'previous_cancellations', \
    'previous_bookings_not_canceled', 'booking_changes', 'adr', 'days_in_waiting_list']
````

````python
n = 1

plt.figure(figsize = (14, 18))

for feature in numeric_features:
    plt.subplot(4,3,n)
    sns.stripplot(x = df['hotel'], y = df[feature]).set(xlabel = None, ylabel = None)
    plt.title(f'{feature}')
    n = n + 1
    plt.tight_layout()
````

On the basis of the category, a scatter plot is created.As shown in plot for the given hotel type many subplot has been created
